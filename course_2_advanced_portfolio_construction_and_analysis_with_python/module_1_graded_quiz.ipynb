```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

from nb import edhec_risk_kit_206 as erk
```

```python
ind = erk.get_ind_returns(weighting="vw", n_inds=49)
fff = erk.get_fff_returns()
```

```python
ind_excess = ind["1991":"2018"] - fff.loc["1991":"2018",['RF']].values
mkt_excess = fff.loc["1991":"2018",['Mkt-RF']]
exp_var = mkt_excess.copy()
exp_var["Constant"] = 1
```

```python
lm = sm.OLS(ind_excess['Beer'], exp_var).fit()
lm.summary()
```

```python
lm = sm.OLS(ind_excess['Steel'], exp_var).fit()
lm.summary()
```

```python
ind_excess = ind["2013":"2018"] - fff.loc["2013":"2018",['RF']].values
mkt_excess = fff.loc["2013":"2018",['Mkt-RF']]
exp_var = mkt_excess.copy()
exp_var["Constant"] = 1
lm = sm.OLS(ind_excess['Beer'], exp_var).fit()
lm.summary()
```

```python
ind_excess = ind["2013":"2018"] - fff.loc["2013":"2018",['RF']].values
mkt_excess = fff.loc["2013":"2018",['Mkt-RF']]
exp_var = mkt_excess.copy()
exp_var["Constant"] = 1
lm = sm.OLS(ind_excess['Steel'], exp_var).fit()
lm.summary()
```

```python
ind_excess = ind["1991":"1993"] - fff.loc["1991":"1993",['RF']].values
mkt_excess = fff.loc["1991":"1993",['Mkt-RF']]
exp_var = mkt_excess.copy()
exp_var["Constant"] = 1

beta = []
industries = ind.columns
for i in industries:
    lm = sm.OLS(ind_excess[i], exp_var).fit()
    beta.append((lm.params[0], i))
beta.sort()
beta
```

```python
ind_excess = ind["1991":"2018"] - fff.loc["1991":"2018",['RF']].values
mkt_excess = fff.loc["1991":"2018",['Mkt-RF']]
exp_var = mkt_excess.copy()
exp_var["Constant"] = 1
exp_var["Value"] = fff.loc["1991":"2018",['HML']]
exp_var["Size"] = fff.loc["1991":"2018",['SMB']]
```

```python
# Size
size = []
industries = ind.columns
for i in industries:
    lm = erk.regress(ind_excess[i], exp_var, alpha=False)
    size.append((lm.params[3], i))
size.sort()
size
```

```python
# Value
value = []
industries = ind.columns
for i in industries:
    lm = erk.regress(ind_excess[i], exp_var, alpha=False)
    value.append((lm.params[2], i))
value.sort()
value
```
